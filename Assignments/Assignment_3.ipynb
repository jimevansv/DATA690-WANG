Read (with naked eyes, not computer) the dictionary and find out the definitions of the following 12 columns and copy the text next to the column name
- UNITID: Unit ID for institution           
- INSTNM:  Institution name        
- CITY: City in which University is located 
- STABBR: State Abbrevation
- ZIP: Zip Code of the location of the University             
- ADM_RATE: Admission Rate in a particular college 
- UGDS: Undergrads Count  
- TUITIONFEE_IN: In Tuition Fee for a particular college  
- MN_EARN_WNE_P10: Mean earnings of students working and not enrolled 10 years after entry       

```python
# reimport the data with above 12 columns (ignore all other variables)
data=pd.read_csv('CollegeScorecard_Raw_Data/MERGED2017_18_PP.csv', usecols = ['UNITID','INSTNM','CITY','STABBR','ZIP',
                                                                              'ADM_RATE','UGDS','TUITIONFEE_IN','MN_EARN_WNE_P10'])
data.info()
```

```python
data.head(5)
```

```python
# Find out the 10 colleges that are hardest to get in (top 10)
## This can be found using the admission rate. The lowest admission rate of any
## college suggests that it is tougher to get in.
toughest_colleges=data.nsmallest(10, ['ADM_RATE']) 
toughest_colleges.head(10)
```

```python
# Below are the names of the colleges
toughest_colleges['INSTNM']
```

```python
# 10. Find out the average in-state tuition of top 10
#toughest_colleges.mean()
toughest_colleges.loc[:,'TUITIONFEE_IN'].mean()
```

```python
# Find out the average earnings of graduates from the top 10
# If we look at the data, all the entires have a null vale (missing data)
print('Average earnings of graduates from the top 10: ')
toughest_colleges.loc[:,'MN_EARN_WNE_P10'].mean()
```

```python
# Find out the 10 colleges that are easiest to get in (bottom 10)
# Easiest colleges mean that the admission rate has to be higher

easiest_colleges=data.nlargest(10, ['ADM_RATE']) 
easiest_colleges.head(10)
```

```python
# Find out the average in-state tuition of the bottom 10
print('Average in-state tuition fee of the bottom 10: ')
easiest_colleges.loc[:,'TUITIONFEE_IN'].mean()
```

```python
# Find out the average earnings of graduates from the bottom 10
# If we look at the data, all the entires have a null vale (missing data)
print('Average earnings of graduates from the bottom 10: ')
easiest_colleges.loc[:,'MN_EARN_WNE_P10'].mean()
```
